```python
from diffusers import DiffusionPipeline
from matplotlib import pyplot as plt
%matplotlib inline
import torch


def test(pipeline):
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    pipeline = pipeline.to(device)

    texts = [
        'A photo of little dog in a bucket', 'A photo of little dog swimming',
        'A photo of little dog sleeping', 'A photo of little dog in a doghouse'
    ]

    images = [pipeline(i).images[0] for i in texts]

    plt.figure(figsize=(10, 5))
    for i in range(4):
        plt.subplot(1, 4, i + 1)
        plt.imshow(images[i])
        plt.axis('off')

    plt.show()
```

```python
pipeline = DiffusionPipeline.from_pretrained('CompVis/stable-diffusion-v1-4',
                                             safety_checker=None)

test(pipeline)
```

```python
pipeline = DiffusionPipeline.from_pretrained('./save', safety_checker=None)

test(pipeline)
```

```python
pipeline = DiffusionPipeline.from_pretrained(
    'lansinuote/diffusion.3.dream_booth', safety_checker=None)

test(pipeline)
```
