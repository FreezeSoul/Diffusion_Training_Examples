```python
from diffusers import DiffusionPipeline
from matplotlib import pyplot as plt
%matplotlib inline
import torch


def test(pipeline):
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    pipeline = pipeline.to(device)

    texts = [
        'a drawing of a green frog with red eyes',#妙蛙种子
        'a green and yellow toy with a red nose',#绿毛虫
        'a red and white ball with an angry look on its face',#雷电球
        'a cartoon character with a potted plant on his head',#椰蛋树
        'a drawing of a dinosaur stuffed animal',#嘎啦嘎啦
        'a drawing of a gray and black dragon',#铁甲犀牛
    ]

    images = [pipeline(i).images[0] for i in texts]

    plt.figure(figsize=(10, 5))
    for i in range(6):
        plt.subplot(2, 3, i + 1)
        plt.imshow(images[i])
        plt.axis('off')

    plt.show()
```

```python
pipeline = DiffusionPipeline.from_pretrained('CompVis/stable-diffusion-v1-4',
                                             safety_checker=None)

test(pipeline)
```

```python
pipeline = DiffusionPipeline.from_pretrained('./save', safety_checker=None)

test(pipeline)
```

```python
pipeline = DiffusionPipeline.from_pretrained(
    'lansinuote/diffusion.4.text_to_image', safety_checker=None)

test(pipeline)
```
