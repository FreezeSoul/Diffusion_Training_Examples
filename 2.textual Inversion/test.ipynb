```python
from diffusers import StableDiffusionPipeline
from matplotlib import pyplot as plt
%matplotlib inline
import torch


def test(pipeline):
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    pipeline = pipeline.to(device)

    texts = [
        'a photo of a <cat-toy>', 'a photo of a clean <cat-toy>',
        'a photo of a dirty <cat-toy>', 'a dark photo of the <cat-toy>'
    ]
    images = []
    for text in texts:
        image = pipeline(text, num_inference_steps=25).images[0]
        images.append(image)

    plt.figure(figsize=(10, 5))
    for i in range(4):
        plt.subplot(1, 4, i + 1)
        plt.imshow(images[i])
        plt.axis('off')

    plt.show()
```

```python
pipeline = StableDiffusionPipeline.from_pretrained(
    'runwayml/stable-diffusion-v1-5')

test(pipeline)
```

```python
pipeline = StableDiffusionPipeline.from_pretrained('./save')

test(pipeline)
```

```python
pipeline = StableDiffusionPipeline.from_pretrained(
    'lansinuote/diffusion.2.textual_inversion')

test(pipeline)
```
